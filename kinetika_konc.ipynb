```python
df=pd.read_csv('/content/кинетика. от конц - Лист1 (2).csv')
df.head()
```

```python
X1=df['t1, c'].values
y1=df['D1'].values
X2=df['t2, c'].values
y2=df['D2']
X3=df['t3,c'].values
y3=df['D3'].values
X4=df['t4,c'].values
y4=df['D4'].values
X5=df['t5,c'].values
y5=df['D5'].values
```

```python
plt.figure(figsize=(13,8))
plt.grid(ls=':')
plt.plot(X1, y1, label='[OH]=0.03М')
plt.plot(X2, y2, label='[OH]=0.012M')
plt.plot(X3,y3, label='[OH]=0.045M')
plt.plot(X4, y4, label='[OH]=0.075M')
plt.plot(X5, y5, label='[OH]=0.1M')
plt.ylabel('Оптическая плотность D')
plt.xlabel('t, c')
plt.legend()
```

```python
a1=np.log(y1)
a2=np.log(y2)
a3=np.log(y3)
a4=np.log(y4)
a5=np.log(y5)
```

```python
plt.figure(figsize=(13, 8))

slope1, intercept1=np.polyfit(np.array(X1), np.array(a1), 1)
slope2, intercept2=np.polyfit(np.array(X2), np.array(a2),1)
slope3, intercept3=np.polyfit(np.array(X3), np.array(a3), 1)
slope4, intercept4=np.polyfit(np.array(X4), np.array(a4), 1)
slope5, intercept5=np.polyfit(np.array(X5), np.array(a5), 1)

plt.plot(X1, a1, label='[OH]=0.03 M, y = {}x + {}'.format("{:.4f}".format(slope1), "{:.4f}".format(intercept1)), color = 'blue')
plt.plot(X2, a2, label='[OH]=0.012 M, y = -0.00582x + {}'.format("{:.4f}".format(slope2), "{:.4f}".format(intercept2)), color = 'orange')
plt.plot(X3, a3, label='[OH]=0.045 M, y = -0.02146x + {}'.format("{:.4f}".format(slope3), "{:.4f}".format(intercept3)), color = 'green')
plt.plot(X4, a4, label='[OH]= 0,075 M, y = -0.037x + {}'.format("{:.4f}".format(intercept4)), color = 'red')
plt.plot(X5, a5, label='[OH]= 0,1 M, y = {}x + {}'.format("{:.4f}".format(slope5), "{:.4f}".format(intercept5)), color = 'purple')

plt.plot(X1, X1*slope1+intercept1, '--', color='blue')
plt.plot(X2, X2*slope2+intercept2, '--', color='orange')
plt.plot(X3, X3*slope3 +intercept3, '--', color='green')
plt.plot(X4, X4*slope4 +intercept4, '--', color='red')
plt.plot(X5, X5*slope5 +intercept5, '--', color='purple')
plt.grid(ls=':')
plt.ylabel('ln(D)')
plt.xlabel('t, c')
plt.legend()
```

```python
data={
    '[OH]=0.03M': [-slope1],
    '[OH]=0.012M': [-slope2],
    '[OH]=0.045M': [-slope3],
    '[OH]=0.075M': [-slope4],
    '[OH]=0.1M': [-slope5]
}
dat=pd.DataFrame(data)
dat
```

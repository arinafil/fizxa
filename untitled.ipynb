 На основе измерений краевого угла от напряжения, вычислим величину поверхностного натяжения и построим зависимость.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
import scipy.stats as sps
import pandas as pd
from scipy import optimize
theta=[82.8, 61.4, 43.2, 37, 32.9, 27.4, 26.1, 23.6, 19.6, 16.3, 16.6,17.4, 11.4, 22.4, 22.4]

sigma=[381.4, 399.41, 412.17, 415.73, 417.82, 420.28, 420.8, 421.73, 423.02, 423.95, 423.87, 423.66, 424.9, 422.15, 422.15]
E=[-1200, -1100, -1000, -900, -800, -700, -600, -500, -400, -300, -200,-100, 0, 100, 200]

theta_1=[82.8, 61.4, 43.2, 37, 17.4, 11.4, 22.4, 22.4]
sigma_1=[381.4, 399.41, 412.17, 415.73,422.15]
E_1=[-1200, -1100, -1000, -900,  200]
```

```python
t=np.polyfit(E, sigma,4)
z=np.polyfit(E_1, sigma_1,2)
f=np.poly1d(t)
g=np.poly1d(z)
x1=[(-1200, 200)]
plt.plot(E, f(E), label='')
plt.plot(E, g(E), '--')
plt.scatter(E, sigma)
plt.grid()
plt.title('$\sigma(E)$')
plt.xlabel('E, mV')
plt.ylabel('$\sigma$, мН/м')
plt.xticks(np.linspace(-1400, 400, 11))
plt.yticks(np.linspace(370, 450, 9))
plt.show()
```

```python
df=pd.read_csv('/content/ВАX.csv')
df.head()
```

```python
X=df['E1'].values
x=X.reshape(-1,1)
Y=df['I1'].values
plt.figure(figsize=(10,6))
plt.plot(x*1000, Y, lw=2)
plt.grid()
plt.title('ВАХ Hg рабочего электрода')
plt.xlabel('E, мВ')
plt.ylabel('I, мкА')
plt.legend()
plt.show()
```

Построим график зависимости тока от концентрации.

```python
c=[1.67,5,8.3,11.6,15,18.3]
i=[1.65,3.22,4.7,6.42,7.62,8.63]#[1.67,3.33,5.5,6.67,9.17,10.83]
C=np.array(c)
I=np.array(i)
plt.scatter(c,i)
model=LinearRegression()
slope, intercept=np.polyfit(C,I,1)
plt.plot(C,C*slope+intercept)
plt.grid()
plt.ylabel('Предельный диффузионный ток,i мкА')
plt.xlabel('C,мМ')
```

```python
print(slope)
```

```python
Ep=[271.41,244.41,230.89,221.05,214,207.96] #мВ
V=[1,3,5,7,9,11]#K4[Fe(CN)6]
V2=1
x=[0,1.1,1.61,1.95,2.2,2.4]
X=np.array(x)
y=np.array(Ep)
plt.figure(figsize=(10,6))
plt.scatter(X,y)
slope,intercept=np.polyfit(X,y,1)
plt.plot(X,X*slope+intercept,lw=1, color='red', label='E=272,35-26,4*X')
plt.grid()
plt.xticks(np.linspace(0,2.5,11))
plt.xlabel('Логарифм отношения концентраций, lnC4/C3', fontsize=12)
plt.ylabel('Стационарный потенциал, мВ', fontsize=14)
plt.legend()
```

```python
c=[0.164,0.268, 0.4346, 0.7134, 1.1644, 1.9, 3.0996, 5.0676, 8.2, 10.25, 13.94, 20.5]
lnc=[-8.716, -8.224, -7.74, -7.245, -6.755, -6.266, -5.776, -5.285, -4.8, -4.6, -4.272, -3.887]
sigma=[61.2,59.9,60.2 , 65.315, 60.365,53.325, 43.215, 38.105, 39.825, 40.1, 40.505,40.4]
plt.scatter(c,sigma)
plt.plot(c, sigma, '--', lw=2, color='cyan')
plt.grid(ls=':')
plt.xlabel('c, мМ', fontsize=12)
plt.ylabel('$\sigma$, мН/м', fontsize=12)

```

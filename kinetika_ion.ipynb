```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('/content/кинетика. от конц - Лист3.csv')
df.head()
```

```python
X1=df['t1'].values
y1=df['D1'].values
X2=df['t2'].values
y2=df['D2']
X3=df['t3'].values
y3=df['D3'].values
X4=df['t4'].values
y4=df['D4'].values
X5=df['t5'].values
y5=df['D5'].values
```

```python
plt.figure(figsize=(13,8))
plt.grid(ls=':')
plt.plot(X1, y1, label='I=0.01M')
plt.plot(X2, y2, label='I=0.03M')
plt.plot(X3,y3, label='I=0.05M')
plt.plot(X4, y4, label='I=0.07M')
plt.plot(X5, y5, label='I=0.1M')
plt.ylabel('Оптическая плотность D')
plt.xlabel('t, c')
plt.legend()
```

```python
a1=np.log(y1)
a2=np.log(y2)
a3=np.log(y3)
a4=np.log(y4)
a5=np.log(y5)
```

```python
plt.figure(figsize=(13, 8))

slope1, intercept1=np.polyfit(np.array(X1), np.array(a1), 1)
slope2, intercept2=np.polyfit(np.array(X2), np.array(a2),1)
slope3, intercept3=np.polyfit(np.array(X3), np.array(a3), 1)
slope4, intercept4=np.polyfit(np.array(X4), np.array(a4), 1)
slope5, intercept5=np.polyfit(np.array(X5), np.array(a5), 1)

plt.plot(X1, a1, label='I=0.01 M, y = -0.0036x + {}'.format("{:.4f}".format(slope1), "{:.4f}".format(intercept1)), color = 'blue')
plt.plot(X2, a2, label='I=0.03 M, y = {}x + {}'.format("{:.4f}".format(slope2), "{:.4f}".format(intercept2)), color = 'orange')
plt.plot(X3, a3, label='I=0.05 M, y = {}x + {}'.format("{:.4f}".format(slope3), "{:.4f}".format(intercept3)), color = 'green')
plt.plot(X4, a4, label='I= 0,07 M, y = {}x + {}'.format("{:.4f}".format(slope4), "{:.4f}".format(intercept4)), color = 'red')

plt.plot(X5, a5, label='I= 0,1 M, y =-0.005 x + {}'.format("{:.4f}".format(slope5), "{:.4f}".format(intercept5)), color = 'purple')

plt.plot(X1, X1*slope1+intercept1, '--', color='blue')
plt.plot(X2, X2*slope2+intercept2, '--', color='orange')
plt.plot(X3, X3*slope3 +intercept3, '--', color='green')
plt.plot(X4, X4*slope4 +intercept4, '--', color='red')
plt.plot(X5, X5*slope5 +intercept5, '--', color='purple')
plt.grid(ls=':')
plt.ylabel('ln(D)')
plt.xlabel('t, c')
plt.legend()
```
